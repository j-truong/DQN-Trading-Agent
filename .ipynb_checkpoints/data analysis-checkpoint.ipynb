### Modelling

```python
from keras import Model
from keras.layers import Input, concatenate, Dense, Flatten
from keras.utils import plot_model
```

```python
window = 10
```

```python
ts_input = Input(shape=(window,), name='TS input')
ts_layer1 = Dense(32, activation='relu', name='ts_layer1')(ts_input)
ts_final = Dense(16, activation='relu', name='ts_layer2')(ts_layer1)
#ts_final = Flatten()(ts_layer2)

bank_input = Input(shape=(1), name='bank input')

concat_layer = concatenate([ts_final, bank_input])
fixed_layer1 = Dense(16, activation='relu', name='fixed_layer1')(concat_layer)
fixed_layer2 = Dense(8, activation='relu', name='fixed_layer2')(fixed_layer1)

action_layer = Dense(4, activation='relu', name='action_layer1')(fixed_layer2)
action_output = Dense(3, activation='linear', name='action_output')(action_layer)

amount_layer = Dense(8, activation='relu', name='amount_layer1')(fixed_layer2)
amount_output = Dense(4, activation='relu', name='amount_output')(amount_layer)

model = Model(inputs=[ts_input, bank_input],
              outputs=[action_output, amount_output])
```

```python
plot_model(model, show_shapes=True)
```
