```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from os import walk

import pickle
from matplotlib.ticker import MultipleLocator

val_l = 14340
```

# Exploration Analysis

```python
fname = "exploration/force_explore.pickle"
with open(fname, 'rb') as handle:
    explore = pickle.load(handle)
    
fname = "exploration/unforce_explore.pickle"
with open(fname, 'rb') as handle:
    unexplore = pickle.load(handle)
```

```python
plt.figure(figsize=(10,4))

val_explore = explore[explore.episode==-1]
val_unexplore = unexplore[unexplore.episode==-1]

val_l = len(val_explore)

plt.plot(range(val_l), val_explore.portfolio, label='explore', linewidth=0.8)
plt.plot(range(val_l), val_unexplore.portfolio, label='unexplore', linewidth=0.8)

#factor = val_explore.portfolio.iloc[0] / val_explore.price.iloc[0]
#plt.plot(range(val_l), factor*val_explore.price, label='buy and hold')

plt.title('validation portfolio \n exploration vs. exploitation')
plt.xlabel('timestep')
plt.ylabel('portfolio')
plt.legend()

plt.savefig('3_exploration')
```

# Reward Analysis

```python
fname = "rewards/reward_positive.pickle"
with open(fname, 'rb') as handle:
    rp = pickle.load(handle)
```

```python
fname = "rewards/reward_percent0.pickle"
with open(fname, 'rb') as handle:
    rp0 = pickle.load(handle)
fname = "rewards/reward_percent1.pickle"
with open(fname, 'rb') as handle:
    rp1 = pickle.load(handle)
fname = "rewards/reward_percent2.pickle"
with open(fname, 'rb') as handle:
    rp2 = pickle.load(handle)
```

```python
plt.figure(figsize=(10,4))
val_rp = rp[rp.episode==-1]

plt.plot(range(val_l), val_explore.portfolio, label='face-value reward', linewidth=0.8)
plt.plot(range(val_l), val_rp.portfolio, label='positive reward', linewidth=0.8)
#plt.plot(range(val_l), factor*val_explore.price, label='buy and hold')

plt.title('validation portfolio \n reward systems')
plt.xlabel('timestep')
plt.ylabel('portfolio')
plt.legend()

plt.savefig('3_reward')
```

```python
plt.figure(figsize=(20,8))
val_rp0 = rp0[rp0.episode==-1]
val_rp1 = rp1[rp1.episode==-1]
val_rp2 = rp2[rp2.episode==-1]

plt.plot(range(val_l), val_rp0.portfolio, label='0.005')
plt.plot(range(val_l), val_rp1.portfolio, label='0.01')
plt.plot(range(val_l), val_rp2.portfolio, label='0.05')

plt.title('validation portfolio \n reward percentage')
plt.xlabel('timestep')
plt.ylabel('portfolio')
plt.legend()
```

# Epsilon Decay

```python
plt.figure(figsize=(10,4))

cols = {'decay', 'ts', 'price', 'inv', 'portfolio', 'action', 'episode', 'bank', 'reward'}
decay_df = pd.DataFrame(columns=cols)
decay_label = '0.9'

for n in range(6):
    fname = 'rl_params/decay'+str(n)+'.pickle'
    with open(fname, 'rb') as handle:
        ed0 = pickle.load(handle)
    
    val_ed0 = ed0[ed0.episode==-1]
    plt.plot(range(val_l), val_ed0.portfolio, label=decay_label, linewidth=0.8)
    
    decay_label += '9'

plt.title('validation portfolio \n epsilon decay')
plt.xlabel('timestep')
plt.ylabel('portfolio')
plt.legend()

plt.savefig('3_epsilon')
```

## DQN ANN

```python
mypath = 'DQN_ANN/train/stats'
_, _, filenames = next(walk(mypath))

plt.figure(figsize=(10,4))

portfolios = pd.DataFrame(columns=['name', 'avg'])

for fname in filenames[1:]:
    with open(mypath+'/'+fname, 'rb') as handle:
        ed0 = pickle.load(handle)
    
    name = fname[:-7]
    val_ed0 = ed0[ed0.episode==-1]
    plt.plot(range(val_l), val_ed0.portfolio, label=name, linewidth=0.8)
    
    portfolios = portfolios.append({'name':name, 'avg':val_ed0.portfolio.mean()}, ignore_index=True)
    
    
plt.title('validation portfolio \n DQN ANN')
plt.xlabel('timestep')
plt.ylabel('portfolio')
plt.legend(loc='lower left', fontsize='x-small')

plt.savefig('5_ANN')
```

```python
# DQN ANN
fname = 'DQN_ANN/train/stats/2_32_32_Adam.pickle'
with open(fname, 'rb') as handle:
    ed0 = pickle.load(handle)

fin = ed0.portfolio[len(ed0)-1]
gain = fin/200
val_len = 1/len(ed0[ed0.episode == -1])

month_roi = (gain**val_len)**28707
month_roi
```

# DQN_CNN

```python
mypath = 'DQN_CNN/train/stats'
_, _, filenames = next(walk(mypath))

plt.figure(figsize=(10,4))

portfolios = pd.DataFrame(columns=['name', 'avg'])

for fname in filenames[1:]:
    with open(mypath+'/'+fname, 'rb') as handle:
        ed0 = pickle.load(handle)
    
    name = fname[:-7]
    val_ed0 = ed0[ed0.episode==-1]
    #val_l = len(val_ed0)
    plt.plot(range(val_l), val_ed0.portfolio, label=name, linewidth=0.8)
    
    portfolios = portfolios.append({'name':name, 'avg':val_ed0.portfolio.mean()}, ignore_index=True)
    

mypath1 = 'DQN_CNN/train/stats/2_32_32_Adam.pickle'
mypath2 = 'DQN_CNN/train/stats/1_32_0_Adam.pickle'
with open(mypath1, 'rb') as handle:
        ed0 = pickle.load(handle)
val_ed0 = ed0[ed0.episode==-1]
plt.plot(range(val_l), val_ed0.portfolio, label=mypath2[:-7])
with open(mypath2, 'rb') as handle:
        ed0 = pickle.load(handle)
val_ed0 = ed0[ed0.episode==-1]
plt.plot(range(val_l), val_ed0.portfolio, label=mypath1[:-7])
    
    
plt.title('validation portfolio \n DQN CNN')
plt.xlabel('timestep')
plt.ylabel('portfolio')
plt.legend(loc='upper left', fontsize='small')
```

```python
mypath = 'DQN_CNN/train/stats'
_, _, filenames = next(walk(mypath))

plt.figure(figsize=(10,4))
portfolios = pd.DataFrame(columns=['name', 'avg'])

model1 = '2_32_32_Adam.pickle'
model2 = '1_32_0_Adam.pickle'

for fname in filenames[1:]:
    with open(mypath+'/'+fname, 'rb') as handle:
        ed0 = pickle.load(handle)
    val_ed0 = ed0[ed0.episode==-1]
    
    if fname==model1:
        plt.plot(range(val_l), val_ed0.portfolio, label=model2[:-7], linewidth=0.8)
    elif fname==model2:
        plt.plot(range(val_l), val_ed0.portfolio, label=model1[:-7], linewidth=0.8)
    else:
        name = fname[:-7]
        plt.plot(range(val_l), val_ed0.portfolio, label=name, linewidth=0.8)
    
    portfolios = portfolios.append({'name':name, 'avg':val_ed0.portfolio.mean()}, ignore_index=True)
    
plt.title('validation portfolio \n DQN CNN')
plt.xlabel('timestep')
plt.ylabel('portfolio')
plt.legend(loc='upper left', fontsize='x-small')

plt.savefig('5_CNN')
```

# DQN_RNN

```python
mypath = 'DQN_RNN/train/stats'
_, _, filenames = next(walk(mypath))

plt.figure(figsize=(10,4))

portfolios = pd.DataFrame(columns=['name', 'value', 'avg'])

for fname in filenames[1:]:
    with open(mypath+'/'+fname, 'rb') as handle:
        ed0 = pickle.load(handle)
    
    name = fname[:-7]
    val_ed0 = ed0[ed0.episode==-1]
    plt.plot(range(val_l), val_ed0.portfolio, label=name, linewidth=0.8)
    
    portfolios = portfolios.append({'name':name, 'value':val_ed0.portfolio[58343], 
                                    'avg':val_ed0.portfolio.mean()}, ignore_index=True)
    
    
plt.title('validation portfolio \n DQN RNN')
plt.xlabel('timestep')
plt.ylabel('portfolio')
plt.legend(loc='lower left', fontsize='x-small')

plt.savefig('5_RNN')
```

## Comparison of all DQN Variations 

```python
plt.figure(figsize=(20,8))

# DQN ANN
fname = 'DQN_ANN/test/stats/2_32_32_Adam.pickle'
with open(fname, 'rb') as handle:
    ed0 = pickle.load(handle)
ts = range(len(ed0))
plt.plot(ts, ed0.portfolio, label='DQN_ANN 2_32_32_Adam')

# DQN CNN
fname = 'DQN_CNN/test/stats/1_32_0_Adam.pickle'
with open(fname, 'rb') as handle:
    ed0 = pickle.load(handle)
plt.plot(ts, ed0.portfolio, label='DQN_CNN 2_32_32_Adam')

# DQN RNN
fname = 'DQN_RNN/test/stats/2_32_8_SGD.pickle'
with open(fname, 'rb') as handle:
    ed0 = pickle.load(handle)
plt.plot(ts, ed0.portfolio, label='DQN_RNN 2_32_8_SGD')

# BUY AND HOLD
factor = ed0.portfolio.iloc[0] / ed0.price.iloc[0]
plt.plot(ts, factor*ed0.price, label='buy and hold')
    

plt.title('test portfolio \n DQN')
plt.xlabel('timestep')
plt.ylabel('portfolio')
plt.legend()
```
